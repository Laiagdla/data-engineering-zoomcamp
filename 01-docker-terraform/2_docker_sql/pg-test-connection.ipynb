````python
df_iter = pd.read_csv('yellow_tripdata_2021-01.csv.gz', iterator=True, chunksize=10000)
````

````python
df_iter
````

````python
next(df_iter)
````

````python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
````

````python
df.head(0).to_sql(name='yellow_tripdata_trip', con=engine, if_exists='replace', index=False)
````

````python
query = """
SELECT * FROM yellow_tripdata_trip LIMIT 10
"""

pd.read_sql(query, con=engine)
````

````python
from time import time

while True:
    try:
        start = time()

        chunk = next(df_iter)
        chunk.tpep_pickup_datetime = pd.to_datetime(chunk.tpep_pickup_datetime)
        chunk.tpep_dropoff_datetime = pd.to_datetime(chunk.tpep_dropoff_datetime)
        chunk.to_sql(name='yellow_tripdata_trip', con=engine, if_exists='append', index=False)
        
        print(f'Chunk loaded in {time() - start} seconds')
    except StopIteration:
        break
````
